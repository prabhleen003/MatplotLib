```python
  # MATPLOTLIB FOR VISUALIZATION OF DATA 
# IT HAS TWO COMPONENTS PYLAB AND PYPLOT 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
#for creating x axis we use linspace 
x = np.linspace(-2,2,9)
y = 2*x*x +3
plt.scatter(x,y)
plt.ylabel('x')
plt.xlabel('y')
plt.title('a scatter plot')
plt.show()
```

```python
# THE POINTS CORRESPONDING TO X MUST BE IN A SORTED FASHION 
plt.plot(x,y)
plt.show()
```

```python
#sixe of the figure can be altered by plt.figure(figuresize())
plt.figure(figsize=(4,4))
plt.plot(x,y)
plt.show()
```

```python
#multiple plotting can be done 
X = np.linspace(-10,10,21)
Y1=X*X
Y2 = X
plt.figure(figsize=(4,4))
plt.plot(X,Y1,label="SQUARE FUNCTION")
plt.plot(X,Y2,label="LINEAR")
plt.ylabel('Y')
plt.xlabel('x')
plt.title("multiple ploting")
plt.legend() # important for showing labels
plt.show()
```

```python
# if the limits of the axes are to be changed 
axes = plt.axes()
# setting bounds
axes.set_xlim([-3,3])
axes.set_ylim([-2,16])
plt.plot(X,Y1, label="poly")
plt.legend(loc="lower left")
plt.show
```

```python
# creating subplots on same display 
x1 = np.linspace(-2,2,9)
y1 = 2*x*x+3
y2 = 2*x
plt.subplot(2,2,1)
plt.scatter(x1,y1,c="red")
plt.subplot(2,2,2)
plt.plot(x1,y1,c="blue")
plt.subplot(2,2,3)
plt.scatter(x1,y2,c="red")
plt.subplot(2,2,4)
plt.plot(x1,y2,c="blue")
plt.show()
```

```python
#piechart
values =[12,55,4,32,14]
colors=['r','b','g','c','m']
explode =[0,0.02,0,0,0]
labels =["a","b","c","d","e"]
plt.pie(values,colors=colors,labels=labels,explode=explode)
plt.title("STUDENTS")
plt.show
```
